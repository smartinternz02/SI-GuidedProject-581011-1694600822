```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Line Graph

```python
df=sns.load_dataset('car_crashes')
```

```python
plt.plot(df['alcohol'],df['abbrev'])
```

```python
plt.plot(df['alcohol'],df['abbrev'],'b->')
```

Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

```python
plt.xticks(rotation=90)
plt.plot(df['abbrev'],df['alcohol'],'b-*',mec = 'r')
plt.grid()
```

### Subplots

```python
x=df['abbrev']
y=df['alcohol']

plt.subplot(2,2,1)
plt.plot(x,y,'b--')
plt.subplot(2,2,2)
plt.plot(x,y,'r*')
plt.subplot(2,2,3)
plt.plot(x,y,'g*')
plt.subplot(2,2,4)
plt.plot(x,y,'r>')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])
#[left,bottom,width,height]
ax1.plot(x,y)

ax2.plot(y,x)
```

## Scatter Plot

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
df.info()
```

```python
df.head(5)
```

```python
sns.scatterplot(x="alcohol",y="not_distracted",data=df)
```

```python
Inference: From the plot we can say that the more the person has drank alcohol more are the chances they are not distracted
```

```python
sns.lineplot(x="not_distracted",y="no_previous",data=df,ci=None)
```

```python
sns.distplot(df["speeding"])
```

```python
sns.relplot(x="ins_losses",y="abbrev",data=df,hue="abbrev")
```

```python
df["abbrev"].value_counts()
```

```python
plt.xticks(rotation = 90)
sns.barplot(data=df,x="speeding",y="alcohol",ci=None)
```

```python
plt.xticks(rotation = 90)
sns.barplot(data=df,x="not_distracted",y="alcohol")
```

```python
plt.xticks(rotation = 90)
sns.countplot(x="total",data=df)
```

```python
sns.jointplot(x="alcohol",y="not_distracted",data=df)
```

```python
plt.xticks(rotation = 90)
sns.boxplot(x="alcohol",y="not_distracted",data=df)
```
